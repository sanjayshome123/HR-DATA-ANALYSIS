# Data Cleaning 



```python
df.dropna(inplace=True) # Drop rows with missing values
df
```

# # Assuming you've loaded your dataset into hr_data DataFrame

# Rename columns

```python


df.rename(columns={
    'Employee_ID': 'ID',
    'Age': 'Age',
    'Gender': 'Sex',
    'Department': 'Dept',
    'Salary': 'Income'
}, inplace=True)

# Check the new column names
print(df.columns)
```

```python
# Check the number of rows before removing duplicates
print("Number of rows before removing duplicates:", len(df))
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Check the number of rows after removing duplicates
print("Number of rows after removing duplicates:", len(df))
```

# Removing Leading/Trailing Whitespaces:

```python
# Assuming 'Dept' column needs sanitization
df['Dept'] = df['Dept'].str.strip()
df
```

# Converting Data Types:

```python
# Assuming 'Age' column needs to be converted to integer
df['Age'] = df['Age'].astype(int)
df
```

# Standardizing Categorical Data:

```python
# Assuming 'Sex' column has values 'male' and 'female', you may want to standardize it to 'Male' and 'Female'
df['Sex'] = df['Sex'].str.capitalize()
df
```

# Ensuring Consistency:

```python
# Assuming 'Sex' column should only have values 'Male' or 'Female', you may want to remove any other values
valid_sex = ['Male', 'Female']
df = df[df['Sex'].isin(valid_sex)]
df
```

```python
# Assuming you want to remove rows with NaN values in the 'Age' and 'DailyRate' columns
df.dropna(subset=['Age', 'DailyRate'], inplace=True)
df
```

```python
# Remove rows with NaN values in any column
df.dropna(inplace=True)
df
```

# additional eda analysis

```python
# Countplot of Attrition
plt.figure(figsize=(6, 4))
sns.countplot(x='Attrition', data=df)
plt.title('Count of Employees by Attrition Status')
plt.show()
```

```python
# Pairplot of numerical variables
sns.pairplot(df[['Age', 'DailyRate', 'MonthlyIncome', 'TotalWorkingYears', 'YearsAtCompany', 'YearsSinceLastPromotion', 'YearsWithCurrManager', 'Attrition']], hue='Attrition', diag_kind='kde')
plt.suptitle('Pairplot of Numerical Variables')
plt.show()
```

```python
# Boxplot of Age by Attrition
plt.figure(figsize=(8, 6))
sns.boxplot(x='Attrition', y='Age', data=df)
plt.title('Age Distribution by Attrition Status')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
```

```python
# Countplot of Department
plt.figure(figsize=(8, 6))
sns.countplot(x='Department', data=df)
plt.title('Count of Employees by Department')
plt.xticks(rotation=45)
plt.show()
```
